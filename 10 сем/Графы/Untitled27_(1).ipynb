```python
import matplotlib.pyplot as plt
```

```python
# Данные
cores = [1, 2, 4, 6, 8, 10, 12, 14, 16]

# Данные для малого графа (из предыдущего замера)
small_rmat = [32.15430, 19.12266, 10.54464, 7.87723, 6.31237, 5.80023, 5.23736, 5.20182, 4.65078]
small_random = [17.32664, 10.98937, 6.07169, 4.82698, 3.76235, 3.62181, 3.49131, 3.33737, 3.17070]

# Данные для большого графа (новые замеры)
large_rmat = [627.43483, 357.10649, 186.78848, 136.06709, 103.84802, 91.79290, 81.13910, 74.74734, 67.02294]
large_random = [559.2834, 312.80011, 162.39394, 118.63255, 89.48482, 79.87974, 69.51612, 63.97404, 56.84808]
```

```python
# Создание области для двух графиков
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --- График 1: Малый размер графа ---
ax1.plot(cores, small_rmat, marker='o', linestyle='-', color='red', label='R-MAT')
ax1.plot(cores, small_random, marker='s', linestyle='-', color='blue', label='Random')
ax1.set_title('Малый граф (Scale 12)', fontsize=14)
ax1.set_xlabel('Число ядер', fontsize=12)
ax1.set_ylabel('Время работы (сек)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_xticks(cores)
ax1.legend()

# --- График 2: Большой размер графа ---
ax2.plot(cores, large_rmat, marker='o', linestyle='-', color='red', label='R-MAT')
ax2.plot(cores, large_random, marker='s', linestyle='-', color='blue', label='Random')
ax2.set_title('Большой граф (Scale 14)', fontsize=14)
ax2.set_xlabel('Число ядер', fontsize=12)
ax2.set_ylabel('Время работы (сек)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_xticks(cores)
ax2.legend()

plt.tight_layout()
plt.show()

# Расчет эффективности для справки
print(f"Ускорение на большом графе (с 2 до 16 ядер):")
print(f"R-MAT:  {large_rmat[0]/large_rmat[-1]:.2f}x")
print(f"Random: {large_random[0]/large_random[-1]:.2f}x")
```
