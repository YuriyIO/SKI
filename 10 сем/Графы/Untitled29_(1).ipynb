```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python

cores = [1, 2, 4, 6, 8, 10, 12, 14, 16]

# Данные RMAT
rmat_no = [627.43, 357.11, 186.79, 136.07, 103.85, 91.79, 81.14, 74.75, 67.02]
rmat_opt = [473.94107, 283.76525, 157.48676, 112.32140, 89.94558, 76.52098, 68.46042, 63.71227, 59.23195]

# Данные Random
rand_no = [559.28, 312.80, 162.39, 118.63, 89.48, 79.88, 69.52, 63.97, 56.85]
rand_opt = [504.65725, 286.55325, 156.31335, 111.95169, 87.37373, 74.48663, 65.09654, 59.95943, 54.83818]
```

```python
# Создание фигуры
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
plt.subplots_adjust(wspace=0.25)

def style_plot(ax, title, data_no, data_opt):
    # Отрисовка линий
    ax.plot(cores, data_no, 'o-', label='Без оптимизации', color='#1f77b4', linewidth=2.5, markersize=8)
    ax.plot(cores, data_opt, 's--', label='RCM Оптимизация', color='#d62728', linewidth=2, markersize=8)

    # Настройка логарифмической шкалы
    ax.set_yscale('log')

    # Форматирование оси Y: превращаем 10^2 в нормальные числа
    ax.yaxis.set_major_formatter(ScalarFormatter())
    ax.yaxis.set_minor_formatter(ScalarFormatter())

    # Задаем конкретные "красивые" точки на оси Y
    y_ticks = [60, 80, 100, 150, 200, 300, 400, 600, 800]
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([str(y) for y in y_ticks])

    # Настройка оси X
    ax.set_xticks(cores)

    # Заголовки и сетка
    ax.set_title(title, fontsize=15, fontweight='bold', pad=15)
    ax.set_ylabel('Время выполнения (сек)', fontsize=12)
    ax.set_xlabel('Число ядер (процессов)', fontsize=12)
    ax.grid(True, which="both", ls="--", alpha=0.6)
    ax.legend(fontsize=11, frameon=True)

# Рисуем оба графика
style_plot(ax1, 'RMAT Граф (Scale 14)', rmat_no, rmat_opt)
style_plot(ax2, 'Random Граф (Scale 14)', rand_no, rand_opt)

plt.suptitle('Сравнение производительности: Базовая версия vs RCM', fontsize=18, y=1.02)
plt.show()
```
