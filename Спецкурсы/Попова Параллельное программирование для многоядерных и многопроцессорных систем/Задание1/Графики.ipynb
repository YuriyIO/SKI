```python
# 1. Устанавливаем размеры матриц и количество потоков
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import time
```

```python
threads_list = [1] + list(range(2, 22, 2))
```

```python
matrixSize = [4000, 4000, 4000]
times = [175.023, 107.076, 60.1912, 44.2214, 31.6426, 26.6961, 16.5402, 14.0564, 12.2872, 10.9184, 9.69127]
t1 = 100

T2 = times[0]  # базовое время на 2 потоках (если нет данных на 1)
speedup = [T2 / t for t in times]
efficiency = [s / p for s, p in zip(speedup, threads_list)]

# === Построение графиков ===
plt.figure(figsize=(18, 5))

# --- 1. Время выполнения ---
plt.subplot(1, 3, 1)
plt.plot(threads_list, times, marker='o', linewidth=2, color='tab:red', label=f"{matrixSize}")
plt.xticks(threads_list)
plt.xlabel("Число потоков", fontsize=12)
plt.ylabel("Время выполнения, с", fontsize=12)
plt.title("Время выполнения в зависимости от числа потоков", fontsize=14)
plt.grid(True)
plt.legend()

# --- 2. Ускорение ---
plt.subplot(1, 3, 2)
plt.plot(threads_list, speedup, marker='o', linewidth=2, color='tab:blue', label=f"{matrixSize}")
plt.xticks(threads_list)
plt.xlabel("Число потоков", fontsize=12)
plt.ylabel("Ускорение", fontsize=12)
plt.title("Ускорение (Speedup)", fontsize=14)
plt.grid(True)
plt.legend()

# --- 3. Эффективность ---
plt.subplot(1, 3, 3)
plt.plot(threads_list, efficiency, marker='o', linewidth=2, color='tab:green', label=f"{matrixSize}")
plt.xticks(threads_list)
plt.xlabel("Число потоков", fontsize=12)
plt.ylabel("Эффективность", fontsize=12)
plt.title("Эффективность (Efficiency)", fontsize=14)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
