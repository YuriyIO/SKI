```python
import matplotlib.pyplot as plt
import numpy as np

procs = np.array([1, 4, 9, 16])
```

```python
read_time = np.array([0.0373408, 0.0443992, 0.0466172, 0.0444462])

compute_time = np.array([0.00819888, 0.00210241, 0.00104436, 0.000609502])

write_time = np.array([0.00120216, 0.00135528, 0.000913421, 0.000840469])

def speedup(time_array):
    return time_array[0] / time_array

def efficiency(time_array, procs):
    return speedup(time_array) / procs

# ============================
# График времени
# ============================
plt.figure(figsize=(10,6))
plt.plot(procs, read_time, marker='o', label='Время чтения')
plt.xlabel('Число процессов')
plt.ylabel('Время (сек)')
plt.title('Время чтения')
plt.xticks(procs)
plt.grid(True)
plt.legend()
plt.show()

# ============================
# График времени
# ============================
plt.figure(figsize=(10,6))
plt.plot(procs, compute_time, marker='s', label='Время умножения')
plt.xlabel('Число процессов')
plt.ylabel('Время (сек)')
plt.title('Время уможения')
plt.xticks(procs)
plt.grid(True)
plt.legend()
plt.show()

# ============================
# График времени
# ============================
plt.figure(figsize=(10,6))
plt.plot(procs, write_time, marker='^', label='Время записи')
plt.xlabel('Число процессов')
plt.ylabel('Время (сек)')
plt.title('Время записи')
plt.xticks(procs)
plt.grid(True)
plt.legend()
plt.show()


# ============================
# График ускорения
# ============================
plt.figure(figsize=(10,6))
plt.plot(procs, speedup(compute_time), marker='o', label='Ускорение умножения')
plt.xlabel('Число процессов')
plt.ylabel('Ускорение')
plt.title('Ускорение')
plt.xticks(procs)
plt.grid(True)
plt.legend()
plt.show()

# ============================
# График эффективности
# ============================
plt.figure(figsize=(10,6))
plt.plot(procs, efficiency(compute_time, procs), marker='s', label='Эффективность умножения')
plt.xlabel('Число процессов')
plt.ylabel('Эффективность')
plt.title('Эффективность')
plt.xticks(procs)
plt.grid(True)
plt.ylim(0,1.1)
plt.legend()
plt.show()
```
