```python
x = [1000, 2000, 4000, 8000, 10000]

x1 = [1000, 2000, 4000, 8000, 10000]
y1 = [5.75094, 11.1163, 22.3334, 43.2631, 54.1194]

# Линия 2
x2 = [1, 2, 3, 4, 5]
y2 = [3.00493, 5.90716, 12.0756, 23.9889, 28.9759]

# Линия 3
x3 = [1, 2, 3, 4, 5]
y3 = [1.60523, 3.2515, 6.2465, 12.4934, 15.9748]

# Подписи
x_label = "Число частиц"
y_label = "Время выполнения (сек)"
title = "Зависимость времени от числа процессов"
# ===========================

# Построение графика
plt.figure(figsize=(8, 5))

plt.plot(x, y1, marker='o', linestyle='-', color='r', label='1 процесс')
plt.plot(x, y2, marker='o', linestyle='-', color='g', label='2 процесса')
plt.plot(x, y3, marker='o', linestyle='-', color='b', label='4 процесса')

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.grid(True)

# Легенда справа
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()  # чтобы график и легенда не накладывались
plt.show()
```

```python
y4 = [5.75094, 11.1163, 22.3334, 43.2631, 54.1194]

y5 = [3.00493, 5.90716, 12.0756, 23.9889, 28.9759]

y6 = [1.60523, 3.2515, 6.2465, 12.4934, 15.9748]

for i in range(0, 5):
  y4[i] = y1[i]/y1[i]
  y5[i] = y1[i]/y2[i]
  y6[i] = y1[i]/y3[i]


# Подписи
x_label = "Число частиц"
y_label = "Ускоренее"
title = "Ускоренее"
# ===========================

# Построение графика
plt.figure(figsize=(8, 5))

plt.plot(x, y4, marker='o', linestyle='-', color='r', label='1 процесс')
plt.plot(x, y5, marker='o', linestyle='-', color='g', label='2 процесса')
plt.plot(x, y6, marker='o', linestyle='-', color='b', label='4 процесса')

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.grid(True)

# Легенда справа
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()  # чтобы график и легенда не накладывались
plt.show()
```

```python
y7 = [5.75094, 11.1163, 22.3334, 43.2631, 54.1194]

y8 = [3.00493, 5.90716, 12.0756, 23.9889, 28.9759]

y9 = [1.60523, 3.2515, 6.2465, 12.4934, 15.9748]

for i in range(0, 5):
  y7[i] = y4[i]/1
  y8[i] = y5[i]/2
  y9[i] = y6[i]/4


# Подписи
x_label = "Число частиц"
y_label = "Эффективность"
title = "Эффективность"
# ===========================

# Построение графика
plt.figure(figsize=(8, 5))

plt.plot(x, y7, marker='o', linestyle='-', color='r', label='1 процесс')
plt.plot(x, y8, marker='o', linestyle='-', color='g', label='2 процесса')
plt.plot(x, y9, marker='o', linestyle='-', color='b', label='4 процесса')

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.grid(True)

# Легенда справа
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()  # чтобы график и легенда не накладывались
plt.show()
```
