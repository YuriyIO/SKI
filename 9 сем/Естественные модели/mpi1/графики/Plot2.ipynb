```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ось X (количество процессов)
x = np.array([1, 2, 4])

# Исходные времена
y1 = [5.75094, 3.00493, 1.60523]
y2 = [22.3334, 12.0756, 6.2465]
y3 = [54.1194, 28.9759, 15.9748]

# Данные и подписи
data = {
    "Частиц 1000": y1,
    "Частиц 4000": y2,
    "Частиц 10000": y3
}

colors = ["skyblue", "orange", "green"]
bar_width = 0.25
offsets = np.arange(-bar_width, bar_width * len(data), bar_width)

# Строим график
for i, (label, y) in enumerate(data.items()):
    plt.bar(x + offsets[i], y, width=bar_width, color=colors[i], edgecolor="black", label=label)

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Время, с")
plt.title("Время выполнения")
plt.xticks(x, x)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
y4 = [y1[0]/t for t in y1]
y5 = [y2[0]/t for t in y2]
y6 = [y3[0]/t for t in y3]

# Строим график ускорения
for i, y in enumerate([y4, y5, y6]):
    plt.bar(x + offsets[i], y, width=bar_width, color=colors[i], edgecolor="black",
            label=["Частиц 1000", "Частиц 4000", "Частиц 10000"][i])

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Ускорение")
plt.title("Ускорение")
plt.xticks(x, x)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
# Вычисляем эффективность
eff4 = [y/x_i for y, x_i in zip(y4, x)]
eff5 = [y/x_i for y, x_i in zip(y5, x)]
eff6 = [y/x_i for y, x_i in zip(y6, x)]

# Строим график эффективности
for i, eff in enumerate([eff4, eff5, eff6]):
    plt.bar(x + offsets[i], eff, width=bar_width, color=colors[i], edgecolor="black",
            label=["Частиц 1000", "Частиц 4000", "Частиц 10000"][i])

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Эффективность")
plt.title("Эффективность")
plt.xticks(x, x)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
