```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
# Данные
x = np.array([1, 2, 4, 6])           # количество процессов
y = [4.12773, 4.39374, 4.79867, 5.34303]  # время выполнения
colors = ["skyblue", "orange", "green", "red"]
N_labels = [1000, 2000, 4000, 6000]

bar_width = 0.5

# Строим столбцы
plt.bar(x, y, width=bar_width, color=colors, edgecolor="black")

# Легенда через Patch
patches = [mpatches.Patch(color=colors[i], label=f"N={N_labels[i]}") for i in range(len(N_labels))]
plt.legend(handles=patches)

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Время выполнения, с")
plt.title("Время выполнения")
plt.xticks(x, x)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
y1_process = [4.12773, 7.79673, 15.9205, 23.7862]

# Вычисляем ускорение
speedup = [y1_process[i] / y[i] for i in range(len(y))]

# Строим столбцы ускорения
plt.bar(x, speedup, width=bar_width, color=colors, edgecolor="black")

# Легенда через Patch
patches = [mpatches.Patch(color=colors[i], label=f"N={N_labels[i]}") for i in range(len(N_labels))]
plt.legend(handles=patches)

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Ускорение")
plt.title("Ускорение")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
# Вычисляем эффективность
efficiency = [s / p for s, p in zip(speedup, x)]

# Строим столбцы эффективности
plt.bar(x, efficiency, width=bar_width, color=colors, edgecolor="black")

# Легенда через Patch
patches = [mpatches.Patch(color=colors[i], label=f"N={N_labels[i]}") for i in range(len(N_labels))]
plt.legend(handles=patches)

# Оформление
plt.xlabel("Количество процессов")
plt.ylabel("Эффективность")
plt.title("Эффективность")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```
