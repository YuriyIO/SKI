```python
import matplotlib.pyplot as plt
```

```python
x = [1, 4, 9, 16]

y = [
     45.3007,
     11.7979,
     6.89343,
     4.80214
 ]

x_label = "Число процессов"
y_label = "Время выполнения (сек)"
title = "Зависимость времени от числа процессов"
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='r')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.xticks(x)
plt.grid(True)
plt.tight_layout()
plt.show()



speedup = [1, 2, 3, 4]
speedup[0] = y[0]/y[0]
speedup[1] = y[0]/y[1]
speedup[2] = y[0]/y[2]
speedup[3] = y[0]/y[3]

x_label = "Число процессов"
y_label = "Ускорение"
title = "Ускорение"
plt.figure(figsize=(8, 5))
plt.plot(x, speedup, marker='o', linestyle='-', color='r')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.xticks(x)
plt.grid(True)
plt.tight_layout()
plt.show()



efficiency = [1, 2, 3, 4]
efficiency[0] = speedup[0]/1
efficiency[1] = speedup[1]/4
efficiency[2] = speedup[2]/9
efficiency[3] = speedup[3]/16

# Подписи
x_label = "Число процессов'"
y_label = "Эффективность"
title = "Эффективность"
# ===========================

# Построение графика
plt.figure(figsize=(8, 5))

plt.plot(x, efficiency, marker='o', linestyle='-', color='r')

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.xticks(x)
plt.grid(True)

plt.tight_layout()  # чтобы график и легенда не накладывались
plt.show()
```
